# **Q1. Handle Missing Values (Basic Cleaning)**

Fill missing values using the following:

Age → Median

Salary → Mean

Hours_Worked_Per_Week → Median

Performance_Score → Mean

Display the dataset after handling missing values.

# **Q7. Build Preprocessing Pipeline**

Create a pipeline that:

Applies encoding to categorical columns

Applies scaling to numerical columns

Use:

ColumnTransformer

Pipeline

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import pandas as pd

# Load dataset
df = pd.read_csv("employee_productivity_dataset.csv")

# Define categorical and numerical columns
categorical_cols = ['Gender', 'Department', 'Work_Mode', 'Location']
numerical_cols = [
    'Age',
    'Salary',
    'Hours_Worked_Per_Week',
    'Projects_Completed',
    'Performance_Score'
]

# Create transformers
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
numerical_transformer = StandardScaler()

# Create ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, categorical_cols),
        ('num', numerical_transformer, numerical_cols)
    ]
)

# Create Pipeline
pipeline = Pipeline(
    steps=[
        ('preprocessor', preprocessor)
    ]
)

# Fit and transform the data
processed_data = pipeline.fit_transform(df)

# Convert output to DataFrame
feature_names = pipeline.named_steps['preprocessor'].get_feature_names_out()

processed_df = pd.DataFrame(
    processed_data,
    columns=feature_names
)

# Display results
print("Processed Dataset Shape:", processed_df.shape)
display(processed_df.head())
```

# **Q8. Apply Pipeline**

Apply the pipeline on the dataset.

Display:

Transformed dataset

Shape of final dataset

```python
X = df[ numerical_cols + categorical_cols]

transformed_data = pipeline.fit_transform(X)

print("Shape of Final Dataset:")
print(transformed_data.shape)

print("\nTransformed Dataset:")
print(transformed_data)
```

# **Q9. Conceptual Question**

Why is scaling important in Python?

# ANSWER

Scaling is important because many machine learning algorithms are sensitive to the magnitude of features.

Benefits:

* Improves model performance.
* Speeds up convergence of algorithms.
* Prevents features with large values from dominating.
* Improves distance calculations in KNN and clustering.
* Helps gradient-based algorithms such as Logistic
* Regression and Neural Networks.

# **Q10. Conceptual Question**

Why do we convert categorical data into numerical form?

# ANSWER

Machine learning algorithms perform mathematical calculations and cannot directly understand text categories.

Benefits:
